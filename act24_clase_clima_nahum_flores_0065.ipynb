#### Breve descripcion el problema
Crear un sistema automatizado en Python (POO) que registre datos meteorológicos de distintas ciudades mediante la entrada del usuario. El programa debe transformar esos datos en información útil, generando automáticamente reportes que incluyan predicciones de estabilidad, alertas por temperaturas extremas y un resumen histórico de las condiciones capturadas.



<img src="https://raw.githubusercontent.com/nahumflores128/Act24-clase-clima-nahum-flores-0065/refs/heads/main/infoclaseclima0065.png" width="600">



```python
class Clima:
    def __init__(self, ciudad, temperatura, humedad, condicion):
        self.ciudad = ciudad
        self.temperatura = temperatura
        self.humedad = humedad
        self.condicion = condicion

    # Función 1: Predecir
    def predecir(self):
        print(f"[PREDICCIÓN]: Para {self.ciudad} se espera que continúe {self.condicion.lower()}.")

    # Función 2: Alertar
    def alertar(self):
        if self.temperatura > 30:
            status = "¡ALERTA! Calor intenso detectado."
        elif self.temperatura < 10:
            status = "¡ALERTA! Temperaturas muy bajas."
        else:
            status = "Clima estable y seguro."
        print(f"[ALERTA]: {status} ({self.temperatura}°C)")

    # Función 3: Historial
    def historial(self):
        print(f"[HISTORIAL]: Ciudad: {self.ciudad} | Humedad: {self.humedad}% | Condición: {self.condicion}")

# --- Entrada de datos para la Ciudad 1 ---
print("--- REGISTRO CIUDAD 1 ---")
c1_nombre = input("Nombre de la ciudad 1: ")
c1_temp = float(input(f"Temperatura en {c1_nombre}: "))
c1_hum = int(input(f"Humedad en {c1_nombre} (%): "))
c1_cond = input(f"Condición (Soleado/Lluvia/etc): ")

# --- Entrada de datos para la Ciudad 2 ---
print("\n--- REGISTRO CIUDAD 2 ---")
c2_nombre = input("Nombre de la ciudad 2: ")
c2_temp = float(input(f"Temperatura en {c2_nombre}: "))
c2_hum = int(input(f"Humedad en {c2_nombre} (%): "))
c2_cond = input(f"Condición (Soleado/Lluvia/etc): ")

# Creamos los dos objetos
ciudad1 = Clima(c1_nombre, c1_temp, c1_hum, c1_cond)
ciudad2 = Clima(c2_nombre, c2_temp, c2_hum, c2_cond)

# --- Ejecución Automática ---

print("\n" + "="*50)
print(f" GENERANDO REPORTES AUTOMÁTICOS ")
print("="*50)

# Reporte para la Ciudad 1
print(f"\n>>> REPORTE DE {ciudad1.ciudad.upper()} <<<")
ciudad1.predecir()
ciudad1.alertar()
ciudad1.historial()

print("-" * 30)

# Reporte para la Ciudad 2
print(f"\n>>> REPORTE DE {ciudad2.ciudad.upper()} <<<")
ciudad2.predecir()
ciudad2.alertar()
ciudad2.historial()

print("\n" + "="*50)
print("SISTEMA FINALIZADO")
print("Programa creado por Nahum NL = 18")
```
